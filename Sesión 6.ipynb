```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("bicicletas-compartidas.csv")
```

```python
df.columns
```

```python
y = df["cuartil-ano"].apply(lambda x: "cuartil-" + str(int(x)))
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel("Cuartiles del año")
ax.set_ylabel("Frecuencia")
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel("Cuartiles del año")
ax.set_ylabel("Frecuencia")
ax.set_facecolor("yellow")
ax.patches[3].set_facecolor("red")
```

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels = y.value_counts().index)
ax.set_title("Pie chart")
```

```python
# Numéricas
y = df["viento"]
```

```python
fig, ax = plt.subplots()
ax.hist(y, bins = 30)
ax.set_title("Histogram")
ax.set_xlabel("Viento")
ax.set_ylabel("Frecuencias")
plt.axvline(np.mean(y), c = "r", linestyle = "--", label = "Mean")
plt.axvline(np.mean(y) +  np.std(y), c = "k", linestyle = "--", label = "+ 1-desviación")
plt.axvline(np.mean(y) - np.std(y), c = "k", linestyle = "--", label = "- 1-desviación")
ax.legend()
```
